```python
import requests
import json
```

```python
base_site = 'https://itunes.apple.com/search'
```

```python
url = base_site + '?term=the+beatles&country=us'
requests.get(url)
```

```python
r = requests.get(base_site, params = {"term": "the beatles", 
                                      "limit": 200,
                                      "country": "us"})
#                                     
r.status_code
```

```python
r.url
```

```python
info = r.json()
print(json.dumps(info, indent = 4))
```

```python
info.keys()
```

```python
print(json.dumps(info['results'][0], indent = 4))
```
